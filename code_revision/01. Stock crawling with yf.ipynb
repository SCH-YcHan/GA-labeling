```python
import pandas as pd
import yfinance as yf
```

```python
NASDAQ = pd.read_csv("../data/Stock_info/NASDAQ.csv")
NASDAQ.sort_values(by=["Market Cap"], ascending=False, inplace=True)
NASDAQ_symbols = NASDAQ[(NASDAQ["IPO Year"]<2009).values & (NASDAQ["IPO Year"].notna()).values & (NASDAQ["Name"].str.contains("Common Stock").values)][0:100]["Symbol"].values
```

```python
NASDAQ_stock = yf.download(NASDAQ_symbols.tolist(), start="2009-01-02", end="2023-11-24")
```

```python
NASDAQ_stock
```

```python
NASDAQ_stock.reset_index().to_csv("../data/Stock_data/NASDAQ.csv", index=False)
```

```python
KOSPI = pd.read_csv("../data/Stock_info/KOSPI.csv", encoding='euc-kr')
KOSPI['종목코드'] = KOSPI['종목코드'].astype(str).str.zfill(6) + ".KS"
KOSPI_symbols = KOSPI.종목코드.values[:100]
```
